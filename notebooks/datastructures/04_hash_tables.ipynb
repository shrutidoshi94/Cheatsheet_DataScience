# Hash Tables (dict / set)
Run the code below to test the Data Structure syntax from the card!

```python
d = {"cat": 5, "dog": 2}
d["fox"] = 9
s = {1, 2, 3}

from collections import defaultdict, Counter
dd = defaultdict(list)
dd['animals'].append('cat')

cnt = Counter("mississippi")

print('Dict:', d)
print('Set:', s)
print('DefaultDict:', dd)
print('Counter:', cnt)
```
